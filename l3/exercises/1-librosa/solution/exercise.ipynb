# Audio Processing with Librosa

This notebook exercises basic audio processing techniques using librosa, including loading files, format conversion, and Mel spectrogram analysis.

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
from IPython.display import Audio

# TODO: load the "music.mp3" audio file
y, sr = librosa.load('music.mp3')

# TODO: load the file at 44.1kHz
y, sr = librosa.load('music.mp3', sr=44100)

# TODO: save as Wav file
import soundfile as sf
sf.write('music.wav', y, sr)

# TODO: plot the waveform
librosa.display.waveshow(y, sr=sr)

# TODO: use the Audio component to play the audio
Audio(y, rate=sr)
```

```python
# TODO: compute and plot the Mel spectrogram
mel_spec = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)

librosa.display.specshow(mel_spec_db, sr=sr, x_axis='time', y_axis='mel')
```

Notice how the beat of the drums is very evident, as well as the phrasing of the guitar in the foreground.
